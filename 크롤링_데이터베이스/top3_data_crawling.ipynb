```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time
import pandas as pd
import urllib.request
import csv
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import os
from datetime import datetime
```

```python
# 현재시간
now = datetime.now()

#크롬 전체화면으로 키기
options = Options()
options.add_argument('--kiosk')
driver = webdriver.Chrome("C:/Users/user/chromedriver/chromedriver.exe", chrome_options=options)

now_date = str(now.date()) + '-' + str(now.hour)
page = 1

# 무신사 url 입력
driver.get("https://store.musinsa.com/app/brand/goods_list/grooverhyme?brand_code=grooverhyme&d_cat_cd=&u_cat_cd=&page_kind=brandshop&list_kind=small&sort=new&sub_sort=&page=1&news_page=1&display_cnt=90&free_dlv=&ex_soldout=&sale_goods=&price=&color=&sex=&tag=&exclusive_yn=&review_total_more=81939&review_total=81941&popup=&blf_yn=&campaign_yn=&like_brand_code=grooverhyme&price1=&price2=&chk_exclusive=&chk_sale=&chk_soldout=") 
# waiting
driver.implicitly_wait(10)
```

```python
while page <= 2:
    if page == 1:
        indexs = list(range(90))
    if page == 2:
        indexs = list(range(90,180))
    if page == 3:
        indexs = list(range(180,270))
    if page == 4:
        indexs = list(range(270,360))
    
    if not os.path.isdir('C:/Users/user/musinsa_store/'):
        os.mkdir('C:/Users/user/musinsa_store/')
    csvFile = open('C:/Users/user/musinsa_store/musinsa_top3_data.csv','a',newline='')
    writer = csv.writer(csvFile)
    
    # 1~2페이지 변경
    pages = page+2
    driver.find_element_by_xpath('//*[@id="product_list"]/div[1]/div[1]/div[2]/div[5]/div/div/a[%s]' %pages).send_keys(Keys.ENTER)
    
    # img 로드될때까지 waiting
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'img[src^="//image.msscdn.net/images/goods_img/"]'))
    )

    # 페이지 스크롤 다운
    target = 1000
    while True:
        now_height = driver.execute_script("return window.pageYOffset;")
        driver.execute_script("window.scrollTo(300, {0});".format(target))
        time.sleep(5)
        target = target + 500
        new_height = driver.execute_script("return window.pageYOffset;")
        if now_height == new_height:
            break

    brands = driver.find_elements_by_xpath('//*[@id="searchList"]/li/div/div[2]/p[1]/a')
    prices = driver.find_elements_by_xpath('//*[@id="searchList"]/li/div/div[2]/p[3]')
    images =[]
    for i in range(0,len(prices)):
        image_xpath = driver.find_elements_by_xpath('//*[@id="searchList"]/li[%s]/div/div[1]/a/img' %(i+1))
        image_src = image_xpath[0].get_attribute('src')
        images.append(image_src)
    
    try:
        brand_l = []
        price_l = []
        result = []

        for brand in brands:
            brand_l.append(brand.text)
        for price in prices:
            price_list = price.text.split("\n")
            if len(price_list)==3:
                price_l.append(price_list[1])
            else:
                price_l.append(price_list[0])

        
        for x in range(0, len(brand_l)):
            result.append(now_date)
            result.append(brand_l[x]+str(indexs[x]))
            result.append(price_l[x])
        
        for y in range(0, len(result),3):
            writer.writerow(result[y:y+3])
        
        for index, image_url in enumerate(images):
            imgdir = 'C:/Users/user/musinsa_store/musinsa_img/top3/%s/' % 'grooverhyme'
            if not os.path.isdir(imgdir):
                os.mkdir(imgdir)
            if page == 2:
                urllib.request.urlretrieve(image_url, imgdir + 'grooverhyme' + str(90+index) + ".jpg")
            else:
                urllib.request.urlretrieve(image_url, imgdir + 'grooverhyme' + str(index) + ".jpg")
        
    finally:
        csvFile.close()
        
    page += 1
    time.sleep(5)

driver.close()
```
